# Comparing two approaches

### First approach 


The first approach was to train one model for recognizing gender and type of clothes in one go. The labels for this model were:


0 - Male trousers 

1 - Male jeans 

2 - Female trousers 

3 - Female jeans 

### Second approach
The second approach was about recognision gender firstly, and then, using one of two classiiers for Trousers and Jeans.

* Gender labels: 

0 - Male

1 - Female

* Jeans and trousers labels:

0 - Trousers

1 - Jeans

# Import libraries

```python
import numpy as np
import tensorflow as tf
```

# Loading the data

```python
data_test  = np.load(r"Data/Trousers & Jeans - All - Test.npz")
```

```python
# Extracting the images array
images_test = data_test['images']

# Extracting the label arrays
gender_test = data_test['genders']
type_test   = data_test['labels']
print("Unique values in type: {}".format(np.unique(type_test[:])))
print("As:\t\t Trousers, Jeans, Trousers, Jeans ")
type_test = type_test%2
print("Unique values in type: {}".format(np.unique(type_test[:])))
```

```python
images_test = images_test/255.0
```

# Loading models


```python
all_labels_dir = r"saved_models\Trousers and jeans classifier\All labels classifier"
gender_dir = r"saved_models\Trousers and jeans classifier\Gender classifier"
male_dir = r"saved_models\Trousers and jeans classifier\Male classifier"
female_dir = r"saved_models\Trousers and jeans classifier\Female classifier"


all_labels_model = tf.keras.models.load_model(all_labels_dir)

gender_model = tf.keras.models.load_model(gender_dir)
male_model = tf.keras.models.load_model(male_dir)
female_model = tf.keras.models.load_model(female_dir)
```

```python
all_labels_model.get_config()
```

# Testing models 

```python
first_approach = 0
second_approach = 0
for i in range(len(images_test)):
    predict_all_labels = all_labels_model(images_test[i:i+1])
    label = np.argmax(predict_all_labels)
    
    #first_approach = first_approach + 1 if label//2 == gender_test[i] else first_approach 
    #first_approach = first_approach + 1 if label%2 == type_test[i] else first_approach
    if label//2 == gender_test[i]:
        first_approach +=1
    if label % 2 == type_test[i]:
        first_approach += 1
    
    predict_gender = gender_model(images_test[i:i+1])
    label_gender  = np.argmax(predict_gender)
    
    if label_gender == gender_test[i]:
        second_approach +=1
    
    if label_gender == 0: 
        clothes_label = male_model(images_test[i:i+1])
    else: 
        clothes_label = female_model(images_test[i:i+1])
    
    label_type = np.argmax(clothes_label)
    if label_type == type_test[i]:
        second_approach +=1
    
    #print("First method points: {}, second method points: {}".format(first_approach, second_approach))
 
```

```python
print("Combining two methods...")
print("Correct prediction for first method: {}".format(first_approach/(2*len(images_test))))
print("Correct prediction for second method: {}".format(second_approach/(2*len(images_test))))
```
